```python
import time
from selenium import webdriver
import pandas as pd

#Download chromedriver https://chromedriver.chromium.org/downloads (https://chromedriver.storage.googleapis.com/index.html?path=79.0.3945.16/)
# Make sure you download the same version as Chrome
# Unzip and save it in the current directory

driver = webdriver.Chrome("/Downloads/chromedriver")
l =[]

# Can add more locations in below list

for loc in ('nc/charlotte/KCLT','nc/morrisville/KRDU','nc/wilmington/KILM','nc/fletcher/KAVL','nc/greensboro/KGSO'):
    for y in range(2019, 2020):
        if y != 2019:
            for m in range(1, 13):
                theurl = "https://www.wunderground.com/calendar/us/"+str(loc)+"/date/"+str(y)+ "-" + str(m)

                driver.get(theurl)
                time.sleep(10)
                for day in driver.find_elements_by_css_selector(".calendar-day.current-month.history"):
                    data = day.text.split('\n')
                    d = {}
                    d["Station"] = loc
                    d["Year"] = y
                    d["Month"] = m
                    d["Day of month"] = (list(data)[0])
                    d["Weather"] = (list(data)[1])
                    d["Temperature"] = (list(data)[3])
                    d["Rain"] = (list(data)[4])
                    l.append(d)
                
        else:
            for m in range(1, 11):
                theurl = "https://www.wunderground.com/calendar/us/"+str(loc)+"/date/"+str(y)+ "-" + str(m)
                driver.get(theurl)
                time.sleep(10)
                for day in driver.find_elements_by_css_selector(".calendar-day.current-month.history"):
                    data = day.text.split('\n')
                    d = {}
                    d["Station"] = loc
                    d["Year"] = y
                    d["Month"] = m
                    d["Day of month"] = (list(data)[0])
                    d["Weather"] = (list(data)[1])
                    d["Temperature"] = (list(data)[3])
                    d["Rain"] = (list(data)[4])
                    l.append(d)
    
#Convert to Dataframe
df = pd.DataFrame(l)
df['Date'] = df['Year'].astype(str) +"-"+ df['Month'].astype(str).str.zfill(2)+"-"+df['Day of month'].astype(str).str.zfill(2)
df['Date']= pd.to_datetime(df['Date']) 
#df.head()

# Export data as a csv file
df.to_csv("/Pastweatherdatafrom2017to2019.csv")

# quits web driver
driver.quit()
```
